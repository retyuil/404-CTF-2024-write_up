```python
from Crypto.Util.number import *
import os
from math import sqrt
flag = b'404CTF{12345678910000000000000000000}'




class LaSeine:
    def __init__(self, size):
        self.l = 20
        self.p = getStrongPrime(size)

        self.a = getRandomNBitInteger(size // 16)
        self.b = getRandomNBitInteger(size // 16)
        print('a = ',self.a)
        print('b = ',self.b)

    def split(self, f):
        if len(f) & 1:
            f += b"\x00"
        h = len(f) // 2
        return (bytes_to_long(f[:h]), bytes_to_long(f[h:]))

    def sign(self, m, b):
        xn, yn = self.split(m)
        k = (getRandomNBitInteger(self.l - 1) << 1) + b
        print('k = ', k)
        for i in range(k):
            xnp1 = (self.a*xn + self.b*yn) % self.p
            ynp1 = (self.b*xn - self.a*yn) % self.p
            if i == 0:
                print('xf1,yf1' ,xnp1,ynp1)
            xn, yn = xnp1, ynp1
        return (k, xn, yn)

seine = LaSeine(1024)

_, xf, yf = seine.sign(flag, 1)
s2 = seine.sign(b"L'eau est vraiment froide par ici (et pas tres propre)", 0)

f = open("out2.txt", "w")

f.write("p = " + str(seine.p) + "\n")
f.write("xf,yf = " + str((xf, yf)) + "\n")
f.write("k,x,y = " + str(s2))

f.close()
f.close()
```

```python
def long_to_bytes(n):
    # Conversion de l'entier long en hexadécimal
    hex_string = "%x" % n
    # Vérification de la longueur de la chaîne hexadécimale
    if len(hex_string) % 2 != 0:
        hex_string = '0' + hex_string  # Ajout d'un zéro initial si nécessaire
    # Conversion hexadécimale en bytes
    byte_string = bytes.fromhex(hex_string)
    return byte_string

for k in range(-2,2):
    
    a2 = a + k
    b = int(sqrt(AB-a2*a2))
    print("A2 : ", a2)
    print('b : ', b)
    xf0 = int((a2*xf1 + b*yf1)/AB)
    print(int(xf0))
    print(long_to_bytes(int(xf0)))
```
